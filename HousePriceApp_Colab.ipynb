```python
# ── 2. Create app directory structure ─────────────────────────────────────────
import os
os.makedirs('app/pages', exist_ok=True)
print('✅ Dossiers créés : app/ et app/pages/')
```

```python
# ── 3. Write app/app.py ────────────────────────────────────────────────────────
app_py = '''
import streamlit as st

st.set_page_config(
    page_title="🏠 House Price ML App",
    page_icon="🏠",
    layout="wide",
    initial_sidebar_state="expanded",
)

st.markdown("""
<style>
  .stApp { background: linear-gradient(135deg, #0f0c29, #302b63, #24243e); }
  [data-testid="stSidebar"] {
      background: linear-gradient(180deg, #1a1a2e 0%, #16213e 50%, #0f3460 100%);
      border-right: 1px solid #e94560;
  }
  [data-testid="stSidebar"] * { color: #e0e0e0 !important; }
  .hero-card {
      background: linear-gradient(135deg, #1a1a2e 0%, #16213e 100%);
      border: 1px solid #e94560; border-radius: 20px;
      padding: 3rem 2rem; text-align: center;
      box-shadow: 0 0 40px rgba(233,69,96,0.3); margin-bottom: 2rem;
  }
  .hero-title {
      font-size: 3.5rem; font-weight: 900;
      background: linear-gradient(90deg, #e94560, #00d4ff, #a855f7);
      -webkit-background-clip: text; -webkit-text-fill-color: transparent;
  }
  h1,h2,h3,h4 { color: #ffffff !important; }
  p, li { color: #c0c0d0 !important; }
</style>
""", unsafe_allow_html=True)

with st.sidebar:
    st.markdown("## 🏠 Navigation")
    st.markdown("---")
    st.markdown("📊 **Page 1** — Données")
    st.markdown("🤖 **Page 2** — Entraînement")
    st.markdown("🔮 **Page 3** — Prédiction")
    st.markdown("---")
    st.info("Ames Housing Dataset\\n\\n• Train : 1 460 lignes\\n• Test  : 1 459 lignes\\n• Cible : SalePrice")

st.markdown("""
<div class=\'hero-card\'>
  <div class=\'hero-title\'>🏠 House Price Predictor</div>
  <div style=\'color:#00d4ff;font-size:1.5rem;font-weight:700\'>ML Application — Ames Housing</div>
  <div style=\'color:#aaa;margin-top:.5rem\'>Explorez · Entraînez · Prédisez</div>
</div>
""", unsafe_allow_html=True)

col1, col2, col3 = st.columns(3)
cards = [
    ("#e94560", "📊", "Page 1 · Données", "Upload CSV, stats, distributions, corrélations"),
    ("#a855f7", "🤖", "Page 2 · Entraînement", "Random Forest, Gradient Boosting, Ridge, Lasso"),
    ("#00d4ff", "🔮", "Page 3 · Prédiction", "Saisie manuelle ou batch CSV"),
]
for col, (color, icon, title, desc) in zip([col1, col2, col3], cards):
    with col:
        st.markdown(f"""
        <div style=\'background:rgba(255,255,255,.04);border:1px solid {color};border-radius:14px;
                    padding:1.5rem;text-align:center;\'>
            <div style=\'font-size:2.5rem\'>{icon}</div>
            <div style=\'color:{color};font-weight:700;font-size:1.1rem;margin:.5rem 0\'>{title}</div>
            <div style=\'color:#aaa;font-size:.9rem\'>{desc}</div>
        </div>""", unsafe_allow_html=True)
'''
with open('app/app.py', 'w') as f:
    f.write(app_py)
print('✅ app/app.py créé')
```

```python
# ── 5. Launch app with ngrok ───────────────────────────────────────────────────
import subprocess, threading, time
from pyngrok import ngrok, conf

# 🔑 Set your ngrok auth token (get one free at https://ngrok.com)
NGROK_TOKEN = "YOUR_NGROK_TOKEN_HERE"  # ← remplacez par votre token ngrok
conf.get_default().auth_token = NGROK_TOKEN

# Kill previous tunnels
ngrok.kill()

# Start Streamlit
def run_streamlit():
    subprocess.run(
        ['streamlit', 'run', 'app/app.py',
         '--server.port', '8501',
         '--server.headless', 'true',
         '--server.enableCORS', 'false'],
        capture_output=False
    )

t = threading.Thread(target=run_streamlit, daemon=True)
t.start()
time.sleep(4)

# Open tunnel
tunnel = ngrok.connect(8501)
print('\n' + '='*60)
print(f'🚀  App accessible ici : {tunnel.public_url}')
print('='*60 + '\n')
```
